__Creating a reusable function for getting  the <i>n</i> most popular <i>M/F</i> Names in the chosen <i>year</i>__

```python
def most_pop(year, gender, n):
    return df.loc[(df.Year == year) & (df.Gender == gender)].nlargest(n, "Count")
```

## 2. Evergreen Names (1880 - 2018)

Objective:
* Finding so-called evergreen Names (the names that are in 1880 and 2018 among the most popular ones) for boys and girls

__Finding evergreen boy Names__ 

```python
# considering popular names as top 30 names 
# (when naming, we need to emphasize that this is variables for male names)
most_pop_M_names_y_1880 = most_pop(1880, "M", 30)
most_pop_M_names_y_2018 = most_pop(2018, "M", 30)
```

__Finding evergreen girl Names__

```python
# considering popular names as top 30 names
# (when naming, we need to emphasize that these is variables for male names)
most_pop_F_names_y_1880 = most_pop(1880, "F", 30)
most_pop_F_names_y_2018 = most_pop(2018, "F", 30)
```

```python
# doing inner merging for identifying the same popular names in chosen years.
# with suffixes straight away
female_evergreen_names = most_pop_F_names_y_1880[["Name", "Count"]].merge(
    most_pop_F_names_y_2018[["Name", "Count"]], 
    how="inner", on="Name", suffixes=("_1880", "_2018"))
female_evergreen_names
```

## 3. Advanced Data Aggregation

Objectives:
* creating a new aggregated DataFrame with some chosen parameters for further exploration
* finding out what is the year with the most counts for each name

```python
# importing undisturbed ds 
df = pd.read_csv("../dataset/us_baby_names.csv")
```

__Creating a new aggregated DataFrame with some chosen parameters for further exploration__

```python
agg = df.groupby(["Name", "Gender"]).agg(Total = ("Count", "sum"), 
                                         No_Year = ("Count", "count"), 
                                         First_Year = ("Year", "min"), 
                                         Last_Year = ("Year", "max"),
                                         Max_Count = ("Count", "max"))
```

```python
agg
```

```python
# finding out what is the year with the most counts for each name. 
# adding to the agg df later on

# defining the proper function
def best_year(group):
    return group.nlargest(1, "Count").Year
```

```python
# applying the best_year() function
best_y = df.groupby(["Name", "Gender"]).apply(best_year)  
```

```python
# running cell above can take a couple of minutes

# now checking the result
best_y  
```

```python
# we are interested in the last column,
# the one that we are going to add to the agg df
agg["Best_Year"] = best_y.droplevel(-1)
```

```python
agg
```

```python
# reseting the index
agg.reset_index(inplace = True)
```

## 6. Creating the "Popularity" and "Rank" Columns

__Objective 2: adding the column "Popularity" - the amount of babies names per million__ 

```python
# building intuition step by step

# getting the number of M and F children by a year 
df.groupby(["Year", "Gender"]).Count.sum()
```

```python
# getting for each name the number of M and F children by a year  
df.groupby(["Year", "Gender"]).Count.transform("sum")
```

```python
# getting the relative popularity in babies per million
df.Count.div(df.groupby(["Year", "Gender"]).Count.transform("sum"))*1000000
```

```python
# finnally, adding the new column 
df["Popularity"] = df.Count.div(df.groupby(["Year", "Gender"]).Count.transform("sum"))*1000000
```

```python
df
```

```python
# looks like we can neglect the digits after the period
pd.options.display.float_format = '{:.0f}'.format
```

```python
df
```

__Objective 2: adding the column "Rank"__

```python
# getting ranks per name per year per gender
df.groupby(["Year", "Gender"]).Count.apply(lambda x: x.rank(ascending=False))
```

```python
# adding the new column 
df["Rank"] = df.groupby(["Year", "Gender"]).Count.apply(lambda x: x.rank(ascending=False))
df
```

## 7. Visualizing Name Trends over Time

Objectives:
* visualization of different features to the Rank column over time for boy/girl Names
* visualization of Popularity and Rank columns over time for evergreen Names

__Visualization of Popularity and Rank over time__

```python
# first, lets create a not reusable plot specifically for the name Mary 
# which represent the correlation between naming popularity and its rank

# separate ds for name Mary  
mary = df[(df.Name == "Mary") & (df.Gender == "F")]
mary
```

```python
# plotting specifically for name Mary
fig, ax1 = plt.subplots(figsize = (12, 8))
ax1.bar(x = mary.Year, height= mary.Popularity, label = "Popularity")
ax2 = ax1.twinx()
ax2.plot(mary.Year, mary.Rank, color = "red", label = "Rank")
ax1.set_xlabel("Year", fontsize = 15)
ax1.set_ylabel("Babies per million", fontsize = 15)
ax2.set_ylabel("Rank", fontsize = 15)
ax1.set_title("Mary", fontsize = 18)
ax1.tick_params(labelsize = 12)
ax2.tick_params(labelsize = 12)
ax1.legend(loc = 2, fontsize = 15)
ax2.legend(loc = 1, fontsize = 15)
plt.show()
```

```python
# we want to make a sort of plotting pattern for later use.
# according to OOP principles, lets create a reusable function for the needed purpose
def plot_name(name, gender, crit = "Popularity"):
    data = df.loc[(df.Name == name) & (df.Gender == gender), ["Name", "Gender", "Year", "Rank", crit]]
    data = data.set_index("Year")
    
    fig, ax1 = plt.subplots(figsize = (12, 8))
    ax1.bar(x = data.index, height= data[crit], label = crit)
    ax2 = ax1.twinx()
    ax2.plot(data.index, data.Rank, color = "red", label = "Rank")
    ax1.set_xlabel("Year", fontsize = 15)
    ax1.set_ylabel(crit, fontsize = 15)
    ax2.set_ylabel("Rank", fontsize = 15)
    ax1.set_title(name, fontsize = 18)
    ax1.tick_params(labelsize = 12)
    ax2.tick_params(labelsize = 12)
    ax1.legend(loc = 2, fontsize = 15)
    ax2.legend(loc = 1, fontsize = 15)
    plt.show()
```

```python
# applying the function

# applying the function for Mary, F
plot_name("Mary", "F")
```

```python
# applying the function for Mary, M
plot_name("Mary", "M")
```

```python
# applying the function for Mary, F, crit = "Count"
plot_name("Mary", "F", crit = "Count")
```

__Visualization of Popularity and Rank over time for evergreen Names__

```python
# for evergreen male names
for name in male_evergreen_names.Name:
    plot_name(name, "M", crit = "Popularity")
```

```python
# for some evergreen male names like John, William, and Joseph, the popularity level is decreasing
# for some like Henry and Samuel, the popularity level tends to recover
# fore some other like David, the popularity level has of bell-alike form

# so perhaps the ones that have decreasing popularity level might be wrong to call as truly evergreen names
```

```python
# for evergreen female names
for name in female_evergreen_names.Name:
    plot_name(name, "F", crit = "Popularity")
```

# 8. Reasoning Name Popularity Sudden Changes

Objectives:
* adding 2 new columns for Popularity one year ago and and for change in popularity from previous year to this particular Year
* reasoning some of positive and negative Name's changes with visualization

__Adding the additional columns for Popularity one year ago and for change in popularity from previous year to this particular Year__  

```python
# before answering the question, let's detect 
# what names have sudden large positive and negative changes in popularity.
# so we need to find data features that caused such a shift.
```

```python
# perhaps the simplest way is to calculate the difference in popularity 
# from one year to another and find the most extreme values

# for instance let's look at name Mary's popularity
df.loc[(df.Name == "Mary") & (df.Gender == "F"), "Popularity"]
```

```python
# adding shift() for comparing with the previous year
df.loc[(df.Name == "Mary") & (df.Gender == "F"), "Popularity"].shift()
```

```python
# by the definition, the very first row has no change compared to the previous year.
# that's why we can see the Nan there

# adding a new column "priv_year_pop_lev" with shifted popularity level 
df["priv_year_pop_lev"] = df.groupby(["Name", "Gender"]).Popularity.shift()

# replacing all NaN-s there with 0
df.loc[(df.Year > 1880) & (df.priv_year_pop_lev.isna()), "priv_year_pop_lev"] = 0
```

```python
df[(df.Name == "Mary") & (df.Gender == "F")]
```

```python
# the next step is adding the column for popularity difference
df["Pop_diff"] = df.Popularity - df.priv_year_pop_lev
```

```python
df[(df.Name == "Mary") & (df.Gender == "F")]
```

__Reasoning some of the positive Name's changes__

```python
# checking the largest twenty most positive differences
df.nlargest(20, "Pop_diff")
```

## 9. Most Popular Unisex Names

Objectives:
* giving a definition of unisex Names 
* find the most popular unisex Names of all times

```python
# two assumptions should be met: 
# 1.to be popular in terms of the total count,
# 2.to have roughly equal spread among boys and girls
```

__Method 1__

```python
# for example, knowing the 1st assumptions is true for the name Mary,
# let's check the 2nd assumption
df[df.Name == "Mary"].groupby(["Name", "Gender"]).Count.sum()
```

```python
# the 2nd assumptions are not quite met

# now, working on all the names

# first, getting the unisex names
unisex = df.groupby(["Name", "Gender"]).Count.sum().unstack()
unisex.dropna(inplace = True)
unisex
```

```python
# second, getting names sorted in high-to-low frequency (in a list form for further location)
unisex_names = unisex.min(axis = 1).sort_values(ascending = False).index
unisex_names
```

```python
# third, locating unisex_names in the unisex ds
unisex.loc[unisex_names].head(20)
```

__Method 2__

```python
#  however, it feels like we can handle this in another way,
# straightly checking the assumptions
```

```python
# 1st assumption checking with threshold popularity of 20000
unisex = unisex[unisex.min(axis = 1) > 20000]
unisex
```

```python
# 2st assumption checking
abs(unisex.F.div(unisex.M).sub(1)).sort_values()
```

```python
# converting to df form and displaying top 20 names
unisex_names2 = abs(unisex.F.div(unisex.M).sub(1)).sort_values().index
unisex.loc[unisex_names2].head(20)
```
